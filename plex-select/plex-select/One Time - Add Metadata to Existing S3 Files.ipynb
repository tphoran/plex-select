```python
s3_resource = boto3.resource('s3')
s3_client = boto3.client('s3')
bucket_name = 'manny-and-meche-plex'
object_summary_iterator = s3_resource.Bucket(bucket_name).objects.all()
for i in object_summary_iterator:
    key = i.key
    metadata = s3_client.head_object(Bucket='manny-and-meche-plex', Key=key)['Metadata']
    if ('Movies/' in key) and \
       (key!='Movies/') and \
       (key[7:-4] not in mismatched) and \
       (('movieid' not in metadata.keys()) or ('proper_movie_title' not in metadata.keys())):
        m = key[7:-4]
        s_result = ia.search_movie(m)
        entry = s_result[0]
        ia.update(entry)
        movieid = entry.movieID
        copy_source = {
            'Bucket': bucket_name,
            'Key': key
        }
        s3_resource.meta.client.copy(copy_source, bucket_name, key,
                                     ExtraArgs={'StorageClass':'ONEZONE_IA',
                                                'Metadata':{'movieID':entry.movieID,
                                                            'proper_movie_title':m}})
        print("Updated "+str(key))
    elif ('Movies/' in key) and \
         (key!='Movies/'):
        print(str(key)+" already processed")
```
